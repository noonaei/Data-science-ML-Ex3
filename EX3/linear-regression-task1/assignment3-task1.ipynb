```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

x= df.dropna()[["age","weight","children","smoker","gender"]]
y= np.array(df.dropna().expenses)

x= pd.get_dummies(data=x.dropna(), drop_first=True)
x = scale.fit_transform(x)
x= np.array(x, dtype= float)

x = sm.add_constant(x)
model = sm.OLS(y ,x)
result = model.fit()

# np.array adds a wierd 1 column to everything so added 1 at start
#second element is age,
#third is wieght(removed BMI for better accurecy)
#forth is children
#fifth is is smoking(answer is no so 0)
#sixth is is male(answer is female so 0)
x_test = np.array([1,66,70.5,1,0,0])
result.predict(x_test)

```
